## Q3>> Compare and contrast mutable and immutable objects in Python with examples.

 Ans.>> Python data type is categorized into two types:

# Mutable Data Type>>
 A mutable data type is those whose values can be changed.
 Example: List, Dictionaries, and Sets

 ### Python Lists
Python lists methods are mutable; the list’s contents be modified in place. You can change a list in place, add new elements, overwrite existing elements, or delete them.
### Python Dictionary
Dictionaries are likely the most important and flexible mutable built-in data types in Python. In simple terms, a Python dictionary is a flexible-sized arbitrary collection of key-value pairs, where key and value both are Python objects.
### Python Sets
Sets are unordered collections of unique elements and immutable objects. You can think of them as a dictionary with just the keys and their values thrown away. Each item of a set should be unique. While sets themselves can be mutable, that is you can add or remove items, however, the items of a Python set must be of an immutable type.
 
# Immutable Data Type>>
 An immutable data type is one in which the values can’t be changed or altered.
 Example: String and Tuples

### Integers (int)
It represents whole numbers, both positive&negative.Once an integer object is created,its value remains fixed.
### Floating Points (float)
It represents real numbers with decimal points,once a floating-point number is initialized, its value remains constant and cannot be altered.
### Strings (str)
Strings are ordered sequences of characters used to represent text.
### Tuples (tuple)
They are ordered collections of elements, similar to lists.
The main distinction between tuples and lists is that tuples are immutable. 
This means that once a tuple is created, its content cannot be changed.
### Booleans (bool)
Booleans represent truth values and can only take True or False values.
These values are constants in Python and cannot be changed.

## Q6>> How do conditional statements work in Python? Illustrate with examples.
Ans.>> 
Conditional Statements are statements in Python that provide a choice for the control flow based on a condition. It means that the control flow of the Python program will be decided based on the outcome of the condition.

Types of Conditional Statements
1. If Conditional Statementn
2. If else Conditional Statementon
3. Nested if..else Conditional Statemeh4n
4. If-elif-else Conditional Statem

# If Conditional Statement in Python
If the simple code of block is to be performed if the condition holds then the if statement is used
# If else Conditional Statements in Python
In a conditional if Statement the additional block of code is merged as an else statement which is performed when if condition is false.
# If-elif-else Conditional Statements in Python
The if statements are executed from the top down. As soon as one of the conditions controlling the if is true, the statement associated with that if is executed, and the rest of the ladder is bypassed. If none of the conditions is true, then the final “else” statement will be executed. .thon

```python
# if statement example
if 10 > 5:
    print("10 greater than 5")

print("Program ended")

print()
# if..else statement example
x = 3
if x == 4:
    print("Yes")
else:
    print("No")

print()
# if-elif statement example
letter = "A"

if letter == "B":
    print("letter is B")

elif letter == "C":
    print("letter is C")

elif letter == "A":
    print("letter is A")

else:
    print("letter isn't A, B or C")
```
